```python
from urllib.request import urlopen
import datetime
from datetime import timedelta
import csv
import pandas as pd
import os
```

```python
#import weather data file

market_start_hour=8
market_finish_hour=14

os.chdir('C:\\Users\\teek2\\Documents\\Farmers Market Forecast\\Weather_Data_Files')
weather_df=pd.read_csv('Los_Angeles_Weather.csv', sep=';'  , engine='python')
```

```python
weather_df
```

```python
grouped_weather_df=weather_df.groupby(['Year','Month','Day']).mean()
```

```python
f1=weather_df['Hour']>=market_start_hour
f2=weather_df['Hour']<=market_finish_hour
weather_df=weather_df[f1 & f2]
```

```python
temps=list(grouped_weather_df['Temperature  [2 m above gnd]'])
```

```python
winds=list(grouped_weather_df['Wind Speed  [10 m above gnd]'])
```

```python
rains=list(grouped_weather_df['Total Precipitation (high resolution)  [sfc]'])
```

```python
dates=sorted([datetime.date(x[0],x[1],x[2]) for x in list(weather_df.groupby(['Year','Month','Day']).groups)])
```

```python
#save weather file
os.chdir('C:\\Users\\teek2\\Documents\\Farmers Market Forecast\\Feature_Data_Files')
rows = zip(dates,temps,winds,rains)
with open('Los_Angeles_Weather_trimmed.csv', "w",newline='') as f:
    writer = csv.writer(f)
    for row in rows:
        writer.writerow(row)
```
